```python
def FeatureSplit(string,Ser=['T1C','T2_FLAIR']):
    import re
    SerUse=None
    for i in sorted(Ser,reverse=True):
        if re.search(i,string)!=None:
            SerUse=i
    ResultDict={'Ser':SerUse,'RawName':string}
    if SerUse!=None:
        temp=string.split('_')[-3:]
        ResultDict.update(dict(zip(['Filter','FeatureGroup','Feature'],temp)))
    else:
        ResultDict.update({'Feature':string})
    return ResultDict
```

```python
import pandas as pd
temp={n:FeatureSplit(s) for n,s in enumerate(['shape_TumorNumber',
                                     '2DT2_FLAIR_Edema_log-sigma-1-mm-3D_glcm_JointEnergy',
                                     '2DT1C_Tumor_log-sigma-1-mm-3D_glrlm_RunVariance'])}
pd.DataFrame(temp).T
```

```python
def Feature2Df(FeatureList):
    import pandas as pd
    temp={n:FeatureSplit(s) for n,s in enumerate(FeatureList)}
    Df=pd.DataFrame(temp).T
    return Df
```
